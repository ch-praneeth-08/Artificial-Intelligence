```python
filt  = df['LanguageWorkedWith'].str.contains('Java', na = False) # why do we use na = False ? because the column has NaN values and we want to ignore them while checking for the condition

df.loc[filt, 'LanguageWorkedWith'] # this will return the data frame with the rows where the LanguageWorkedWith column contains 'Java'
```
